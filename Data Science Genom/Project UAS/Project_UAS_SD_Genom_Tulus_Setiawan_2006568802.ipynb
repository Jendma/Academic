# Differential Expression Gene (4 subtype cancer)

Akan dilakukan analisis data genetik dengan fokus pada identifikasi gen yang menunjukkan perbedaan ekspresi antara empat kelompok sub tipe kanker, yaitu kelompok ependymoma, glioblastoma, medulloblastoma, dan pilocytic_astrocytoma. Proses analisis ini menggunakan teknik library Limma. Langkah awal yang dilakukan adalah dengan membuat grup sub tipe kanker, yang terdiri dari
empat kelompok tersebut.


```r
vargrp <- subset_df[,2]
table(vargrp)
```

```r
subtype_cancer_idx <- ifelse(vargrp %in% c('ependymoma', 'glioblastoma', 'medulloblastoma', 'pilocytic_astrocytoma'),
                         0, 1)
subtype_cancer <- vargrp[subtype_cancer_idx == 0]

group_subtype_cancer <- ifelse(subtype_cancer == "ependymoma", 0,
                       ifelse(subtype_cancer == "glioblastoma", 1,
                              ifelse(subtype_cancer == "medulloblastoma", 2, 3)))
cat("Kategori sub-cancer:\n")
subtype_cancer
cat("\n")
cat("Grup sub-cancer:\n")
group_subtype_cancer
```

```r
expdtgeo_sc = expdtgeoFilt[, subtype_cancer_idx==0]
head(expdtgeo_sc)
```

**LIMMA ANALYSIS**

Selanjutnya, analisis menggunakan paket Limma, yaitu suatu modifikasi model linear yang digunakan khususnya untuk menganalisis ekspresi gen atau protein. Library Limma digunakan untuk memperoleh informasi statistik seperti F-statistic, p-value, dan koefisien yang terkait dengan masing-masing gen. Proses dimulai dengan pembuatan model desain linear yang melibatkan variabel kelompok yang ingin dibandingkan, yaitu yang bersubtipe kanker saja. Hasilnya, kita mendapatkan koefisien untuk setiap gen, yang mencerminkan perbedaan ekspresi antarsubtipe. Informasi
tambahan, seperti p-value dan F-statistic, memberikan dasar statistik untuk menentukan signifikansi perbedaan tersebut.

```r
BiocManager::install('limma')
```

```r
library(limma)
design <- model.matrix(~group_subtype_cancer)
fit <- eBayes(lmFit(expdtgeo_sc,design))
fit
```

Volcano plot di bawah merepresentasikan bahwa kebanyakan dari selisih ekspresi gen dari keempat kelompok sub tipe kanker cenderung tidak terlalu besar dibuktikan dengan
sebaran titik dari volcano plotnya berada di interval antara -0.2 hingga 0.4 meskipun p-value yang dihasilkan sudah signifikan. Hasil tersebut dapat kita abaikan karena
bukan merupakan bagian yang penting.


```r
volcanoplot(fit, coef=2)
```

**INTERPRETASI DARI 20 GEN TERATAS**

```r
topResult <- topTable(fit, coef=2, number=20)

rownames(topResult)
```

```r
selected <- rownames(expdtgeo_sc) %in% rownames(topResult)
expdtgeosel <- expdtgeo_sc [selected, ]

heatmap(expdtgeosel)
```

Penggunaan heatmap membantu memvisualisasikan pola ekspresi dari 20 gen teratas. Heatmap berikut menggambarkan expresi dari setiap gen (baris) dan setiap
sampel (kolom). Warna merah menandakan gen tersebut memiliki ekspresi lebih tinggi (*over expression*). Terlihat pengelompokkan yang jelas antara sampel dari keempat
grup, yaitu ependymoma, glioblastoma, medulloblastoma, dan pilocytic_astrocytoma.


```r
par(mfrow=c(2,2))
for (i in 1:4) plot(group_subtype_cancer, expdtgeosel[i,],
    main = rownames(expdtgeosel)[i])

cat("Keterangan untuk axis group_subtype_cancer:\n
     0.0: ependymoma\n
     1.0: glioblastoma\n
     2.0: medulloblastoma\n
     3.0: pilocytic_astrocytoma")
```

## Feature Selection menggunakan lasso regression

```r
BiocManager::install("caret")
```

```r
BiocManager::install("glmnet")
```

Load R package yang dibutuhkan:
1. tidyverse untuk memudahkan manipulasi dan visualisasi data
2. caret untuk memudahkan workflow machine learning
3. glmnet untuk untuk menghitung penalized regression


```r
library(tidyverse)
library(caret)
library(glmnet)
```

```r
train_index <- sample(nrow(t(X)), 0.8 * nrow(t(X))) # create index for training set
train_X <- t(X)[train_index, ] # create training set
train_y <- as.factor(y[train_index]) # create training labels

test_X <- t(X)[-train_index, ] # create test set
test_y <- as.factor(y[-train_index]) # create test labels
```

```r
train <- cbind(train_X, train_y)
test <- cbind(test_X, test_y)
```

Pisahkan data menjadi data training dan data testing

```r
dim(train)
dim(test)
```

```r
head(train)
head(test)
```

```r
lambda <- 10^seq(-2, 2, length = 1000)
lambda
```

```r
set.seed(2006568802)
lasso <- train(
  train_X,train_y, method = "glmnet",
  trControl = trainControl("cv", number = 5),
  tuneGrid = expand.grid(alpha = 1, lambda = lambda)
  )
# Model coefficients
coef(lasso$finalModel, lasso$bestTune$lambda)
```

Untuk mendapatkan prediktor dan koefisien model "terbaik".
Tanda . menunjukkan bahwa koefisiennya adalah 0

```r
library(Matrix)
non_zero_indices <- which(coef(lasso$finalModel, lasso$bestTune$lambda) != 0, arr.ind = TRUE)
non_zero_values <- coef(lasso$finalModel, lasso$bestTune$lambda)[non_zero_indices]

# Get row and column names (if they exist)
rows <- rownames(coef(lasso$finalModel, lasso$bestTune$lambda))[non_zero_indices[, 1]]
cols <- colnames(coef(lasso$finalModel, lasso$bestTune$lambda))[non_zero_indices[, 2]]

# Combine into a data frame for easier viewing/manipulation
coef_selected <- data.frame(Row = rows, Column = cols, Value = non_zero_values)
coef_selected
```

```r
lasso$bestTune
```

## Model klasifikasi dengan logistic regression

```r
feature_selected <- c(rows[2:length(rows)])
X_selected <- train_X[, feature_selected]
head(X_selected)
```

```r
# Fit the final model on the training data
model <- glmnet(X_selected, train_y, alpha = 0, lambda = 0, family='binomial')
# Display regression coefficients
coef(model)
```

```r
test_X_selected <- test_X[, feature_selected]
```

```r
probabilities <- model %>% predict(newx = test_X_selected)
predicted.classes <- ifelse(probabilities > 0.5, 1, 0)
# Model accuracy
observed.classes <- test_y
mean(predicted.classes == observed.classes)
```

## Model klasifikasi dengan XGBoost

```r
feature_selected <- c(rows[2:length(rows)])
train_X_selected <- train_X[, feature_selected]
test_X_selected <- test_X[, feature_selected]
train_X <- data.matrix(train_X_selected)
test_X <- data.matrix(test_X_selected)
```

```r
BiocManager::install('xgboost')
```

```r
# Load the xgboost package
library(xgboost)
```

```r
# Convert the data into xgb.DMatrix objects
train_matrix <- xgb.DMatrix(data = train_X, label = as.vector(train_y))
test_matrix <- xgb.DMatrix(data = test_X, label = as.vector(test_y))
```

```r
# Specify the parameters for the XGBoost model
params <- list(
  objective = "binary:logistic", # binaryclass classification
  max_depth = 3, # maximum depth of each tree
  subsample = 0.8, # subsample ratio of the training data
  colsample_bytree = 0.8 # subsample ratio of the features
)
```

```r
# Train the model
model <- xgb.train(params = params, data = train_matrix, nrounds = 100)
```

```r
# Make predictions on the test data
preds <- predict(model, newdata = test_matrix)
```

```r
predicted.classes <- ifelse(preds >= 0.5, 1, 0)
observed.classes <- test_y
mean(predicted.classes == observed.classes)
```
